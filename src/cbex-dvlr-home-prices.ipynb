```python
import pandas as var_pandas
import numpy as var_numpy
import matplotlib.pyplot as var_pyplot
import math as var_math
from sklearn import linear_model as var_linear_model
from sklearn import preprocessing as var_preprocessing
from sklearn import compose as var_compose
```

```python
train_df = var_pandas.read_csv("../data/cbex-dvlr-home-prices-train.csv")
train_df
```

```python
dummies = var_pandas.get_dummies(train_df.locality)
dummies
```

```python
dummied_train_df = var_pandas.concat([train_df,dummies],axis='columns')
dummied_train_df
```

```python
dummied_train_df = dummied_train_df.drop(['locality', 'bangalore'], axis='columns')
dummied_train_df
```

```python
dummied_model_one = var_linear_model.LinearRegression()
dummied_model_one.fit(dummied_train_df[['area','chennai','hyderabad']], dummied_train_df.price)
dummied_model_one
```

```python
dummied_train_inputs = dummied_train_df.drop(['price'], axis='columns')
dummied_train_inputs
```

```python
dummied_train_outputs = dummied_train_df.price
dummied_train_outputs
```

```python
dummied_model_two = var_linear_model.LinearRegression()
dummied_model_two.fit(X=dummied_train_inputs, y=dummied_train_outputs)
dummied_model_two
```

```python
oh_encoder = var_preprocessing.OneHotEncoder()
column_transformer = var_compose.make_column_transformer((oh_encoder, ['locality']), remainder='passthrough')

oh_encoded_train_df = train_df
transformed_data = column_transformer.fit_transform(oh_encoded_train_df)
transformed_data
```

```python
oh_encoded_train_df = var_pandas.DataFrame(transformed_data, columns=column_transformer.get_feature_names_out())
oh_encoded_train_df
```

```python
oh_encoded_train_df = oh_encoded_train_df.rename( columns = {
    "onehotencoder__locality_bangalore" : "bangalore", 
    "onehotencoder__locality_hyderabad" : "hyderabad", 
    "onehotencoder__locality_chennai"   : "chennai",
    "remainder__area" : "area",
    "remainder__price" : "price"}
)
oh_encoded_train_df
```

```python
oh_encoded_train_df = oh_encoded_train_df.drop(['bangalore'], axis='columns')
oh_encoded_train_df
```

```python
oh_encoded_train_df = oh_encoded_train_df.reindex(columns=['area','price','chennai','hyderabad'])
oh_encoded_train_df
```

```python
oh_encoded_train_inputs = oh_encoded_train_df.drop(['price'], axis='columns')
oh_encoded_train_inputs
```

```python
oh_encoded_train_outputs = oh_encoded_train_df.price
oh_encoded_train_outputs
```

```python
oh_model = var_linear_model.LinearRegression()
oh_model.fit(X=oh_encoded_train_inputs, y=oh_encoded_train_outputs)
oh_model
```

```python
dummied_model_one_score = dummied_model_one.score(X=dummied_train_inputs, y=dummied_train_outputs)
dummied_model_two_score = dummied_model_two.score(X=dummied_train_inputs, y=dummied_train_outputs)
oh_model_score = oh_model.score(X=oh_encoded_train_inputs, y=oh_encoded_train_outputs)
```

```python
print ("=== Trained Model Outuputs ===")
print ("Dummy-1: Coefficients = {}, Intercept = {}, Score = {}".format(dummied_model_one.coef_, dummied_model_one.intercept_, dummied_model_one_score))
print ("Dummy-2: Coefficients = {}, Intercept = {}, Score = {}".format(dummied_model_two.coef_, dummied_model_two.intercept_, dummied_model_two_score))
print ("One-Hot: Coefficients = {}, Intercept = {}, Score = {}".format(oh_model.coef_, oh_model.intercept_, oh_model_score))
```

```python
test_df = var_pandas.read_csv("../data/cbex-dvlr-home-prices-test.csv")
test_df.head(5)
```

```python
predicted_prices_dummied_one = dummied_model_one.predict(test_df)
predicted_prices_dummied_two = dummied_model_two.predict(test_df)
predicted_prices_oh = oh_model.predict(test_df)

test_df['predicted_price_dummied_one'] = predicted_prices_dummied_one
test_df['predicted_price_dummied_two'] = predicted_prices_dummied_two
test_df['predicted_price_oh'] = predicted_prices_oh
test_df.to_csv("../.outputs/.predictions/cbex-dvlr-home-prices-predictions.csv", index=False)
test_df
```
